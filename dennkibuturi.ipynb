```python
import math
def eedf(e):
    pi = math.pi
    return ((4*(e/pi))**(1/2))*(math.exp(-e))
def deedf(e):
    return ((0.565)*(e)**(1/2))*(math.exp(-0.243*(e)**2)) 
```

```python
import matplotlib.pyplot as plt

plt.rcParams['legend.fontsize'] = 16
plt.figure(figsize=(10,7))

x = [i/100 for i in range(1000)]
ym = []
yd = []
for a in x:
    ym.append(eedf(a))
    yd.append(deedf(a))


plt.plot(x, ym, color = "green", label="Maxwellian")
plt.plot(x, yd, color = "blue", label="Druyvesteyn")


plt.tick_params(labelsize=15)
plt.xlabel("Electron Energy [eV]", fontsize=18)
plt.ylabel("Probability", fontsize=18)
plt.legend()
plt.grid(linestyle='dotted',)
```
