```python
import math
def pi(a,b):
    return (math.exp(-a)-b)/(1-b)
```

```python
import matplotlib.pyplot as plt

x = [i/100 for i in range(1000)]
y = []
for xi in x:
    y.append(pi(xi, 0.2))

x2 = [i/100 for i in range(1000)]
y2 = []
for xi in x2:
    y2.append(pi(xi, 0.1))

x3 = [i/100 for i in range(1000)]
y3 = []
for xi in x3:
    y3.append(pi(xi, 0.05))

x4 = [i/100 for i in range(1000)]
y4 = []
for xi in x4:
    y4.append(pi(xi, 0.025))

plt.figure(figsize=(10,7))
plt.tick_params(labelsize=15)
plt.yscale("log")
plt.xlabel("Δv/Ue",fontsize=18)
plt.ylabel("π",fontsize=18)
plt.plot(x, y, color = "green")
plt.plot(x2, y2, color = "red")
plt.plot(x3, y3, color = "blue")
plt.plot(x4, y4, color = "yellow")
```
