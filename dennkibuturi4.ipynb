```python
# 行列の式を記述
def a1(v1,v2):
    return 2*v1 - v2 -1

def a2(v1,v2,v3):
    return v1 - (-5/2)*v2 + v3

def a3(v2,v3):
    return v2 - (4/3)*v3
```

```python
def f1(V2):
    return (1/2)*(1+V2)

def f2(V1,V3):
    return (-2/5)*(-V1-V3)

def f3(V2):
    return (-3/4)*(-V2)
```

```python
import math
# ガウス・ヤコビ緩和法
def gauss_jacobi(j1, j2, j3, eps):
    for i in range(100):
    
        v1 = f1(j2)
        v2 = f2(j1,j3)
        v3 = f3(j2)
        if math.fabs(a1(v1,v2)) < eps and math.fabs(a2(v1,v2,v3)) < eps and math.fabs(a3(v2,v3)) < eps:
            break
        j1 = v1     # 新しい値の代入
        j2 = v2     # 新しい値の代入
        j3 = v3     # 新しい値の代入
        print(i,"回目 : ", 'v1: {:.4f}'.format(v1),' v2: {:.4f}'.format(v2),' v3: {:.4f}'.format(v3))
```

```python
gauss_jacobi(0,0,0,1e-5)
```

```python
import math
# ガウス・ザイデル緩和法
def gauss_zaidel(j1, j2, j3, eps):
    for i in range(100):
    
        v1 = f1(j2)
        v2 = f2(v1,j3)
        v3 = f3(v2)
        if math.fabs(a1(v1,v2)) < eps and math.fabs(a2(v1,v2,v3)) < eps and math.fabs(a3(v2,v3)) < eps:
            break
        j2 = v2     # 新しい値の代入
        j3 = v3     # 新しい値の代入
        print(i,"回目 : ", 'v1: {:.4f}'.format(v1),' v2: {:.4f}'.format(v2),' v3: {:.4f}'.format(v3))
```

```python
gauss_zaidel(0,0,0,1e-5)
```
