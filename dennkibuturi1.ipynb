```python
import math
def eedf(e, k,T):
    pi = math.pi
    return ((4*e/pi)**(1/2))*((1/(k*T))**(3/2))*(math.exp(-e/(k*T))) 
```

```python
import matplotlib.pyplot as plt
x = [i/100 for i in range(1000)]
y = []
for a in x:
    y.append(eedf(a, 1, 1))


plt.figure(figsize=(10,7))
plt.tick_params(labelsize=15)
plt.xlabel("X-LABEL")
plt.ylabel("Y-LABEL")
plt.plot(x, y, color = "green")
```
